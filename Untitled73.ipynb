```python
def transformers_embedding_model(doc_id, num_results):
    # doc_id: apparel's id in given corpus
    
    # pairwise_dist will store the distance from given input apparel to all remaining apparels
    # the metric we used here is cosine, the coside distance is mesured as K(X, Y) = <X, Y> / (||X||*||Y||)
    # http://scikit-learn.org/stable/modules/metrics.html#cosine-similarity
    print('='*100)
    print("ASIN from dataset 1:",data_1.iloc[doc_id]['asin'])
    print("Title from dataset 1:",data_1.iloc[doc_id]['title'])
    print("Brand from dataset 1:",data_1.iloc[doc_id]['brand'])
    print('='*100)
    pairwise_dist = cosine_similarity([sentence_embeddings_1[doc_id]],sentence_embeddings_2)
    
    # np.argsort will return indices of the smallest distances
    indices = np.argsort(pairwise_dist.flatten())[::-1][0:num_results]
    #pdists will store the smallest distances
    pdists  = np.sort(pairwise_dist.flatten())[::-1][0:num_results]

    #data frame indices of the 9 smallest distace's
    df_indices = list(data_2.index[indices])
    
    for i in range(0,len(indices)):
        # we will pass 1. doc_id, 2. title1, 3. title2, url, model
        #get_result(indices[i],data['title'].loc[df_indices[0]], data['title'].loc[df_indices[i]], data['medium_image_url'].loc[df_indices[i]], 'bag_of_words')
        print('ASIN :',data_2['asin'].loc[df_indices[i]])
        print('.'*100)
        print ('Brand:', data_2['brand'].loc[df_indices[i]])
        print('.'*100)
        print ('Title:', data_2['title'].loc[df_indices[i]])
        print('.'*100)
        print ('Cosine similarity with the query point',pdists[i])
        print('='*100)
```

```python
transformers_embedding_model(19,5)
```
